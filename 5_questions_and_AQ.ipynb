```python
subscores = ''' social skill (items 1,11,13,15,22,36,44,45,
47,48); attention switching (items 2,4,10,16,25,32,34,
37,43,46); attention to detail (items 5,6,9,12,19,23,28,
29,30,49); communication (items 7,17,18,26,27,31,33,
35,38,39); imagination (items 3,8,14,20,21,24,40,41,
42,50)'''
```

```python
subscores_dict = {}

for line in subscores.replace('\n','').strip().replace(' ','_').split(';'):
    if line.startswith('_'):
        line = line[1:]
    if line.endswith(')'):
        line = line[:-1]
    k,v = line.split('_(items_')
    subscores_dict[k] = [int(i) for i in v.split(',')]
```

```python
scoring = {'agree': [1, 2, 4, 5, 6, 7,
9, 12, 13, 16, 18, 19, 20, 21, 22, 23, 26, 33, 35, 39, 41,
42, 43, 45, 46],
'disagree':
[3, 8, 10, 11, 14, 15, 17, 24, 25, 27, 28, 29, 30, 31, 32,
34, 36, 37, 38, 40, 44, 47, 48, 49, 50]}
```

```python
scoring['agree']
```

```python
translator = {n+1:c for n,c in enumerate(aq.columns)}
```

```python
aq.columns = translator.keys()
```

```python
def agree_score(x):
    return [1 if i in {'Slightly agree', 'Definitely agree'} else 0 for i in x]
```

```python
aq[scoring['agree']] = aq[scoring['agree']].apply(agree_score)
```

```python
def disagree_score(x):
    return [1 if i in {'Slightly disagree', 'Definitely disagree'} else 0 for i in x]
```

```python
aq[scoring['disagree']] = aq[scoring['disagree']].apply(disagree_score)
```

```python
sns.distplot(aq.sum(axis = 'columns'))
```

```python
for k,v in subscores_dict.items():
    print(k,v)
    aq[k] = aq[v].sum(axis = 'columns')
```
